```python
data = data_df.sample(3500)
```

```python
#Prompt

data['text'] = data[['dialogue','summary']].apply(lambda x : "Human: Summarize the following text: " + x['dialogue'] + "\nAssistant: " + x["summary"], axis=1)
```

```python
data = Dataset.from_pandas(data)
```

##Using Unsloth

```python
from unsloth import FastLanguageModel
from trl import SFTTrainer
from transformers import TrainingArguments, AutoModelForCausalLM, AutoTokenizer
```

```python
model,tokenizer = FastLanguageModel.from_pretrained(
    model_name = 'unsloth/mistral-7b-instruct-v0.2-bnb-4bit',
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = True)
```

```python
print(model)
print(tokenizer)
```

## LoRA with Quantized Model

```python
model_unsloth = FastLanguageModel.get_peft_model(
    model,
    r=16,
    lora_alpha=16,
    lora_dropout=0,
    target_modules=['q_proj',"k_proj","v_proj","o_proj","gate_proj",'up_proj',"down_proj"],
    bias="none",
    use_gradient_checkpointing=True,
    random_state=42,
    max_seq_length=2048
)
```

```python
training_arg = TrainingArguments(
    per_device_train_batch_size=8,
    gradient_accumulation_steps=1,
        report_to="none",
    output_dir = 'content/unsloth_mistral-4bit',
    warmup_steps=20,
    max_steps=100,
    learning_rate=2e-4,
    fp16=not torch.cuda.is_bf16_supported(),
    bf16=torch.cuda.is_bf16_supported(),
    optim="adamw_8bit",
    seed = 500,
    eval_steps=10,  # Evaluate every 10 steps
    logging_steps=10,
)
```

```python
trainer = SFTTrainer(
    model = model_unsloth,
    train_dataset = data,
    dataset_text_field = 'text',
    tokenizer = tokenizer,
    max_seq_length = 2048,
    args = training_arg,
    )
```

```python
#@title Show current memory stats
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
```

```python
stats = trainer.train()
```

## Final Memory & Time Stamp

```python
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory         /max_memory*100, 3)
lora_percentage = round(used_memory_for_lora/max_memory*100, 3)
print(f"{stats.metrics['train_runtime']} seconds used for training.")
print(f"{round(stats.metrics['train_runtime']/60, 2)} minutes used for training.")
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
```
